```python
subjIDs = utils.getSubjIDs()
subjID = subjIDs[36]
task = utils.taskList[0]
df = utils.load_mouseMovement(subjID, task)
# df.head()
```

```python
wSize = int(1.2 * 60)
columns = ['x-shift', 'y-shift']
nTraj = 50
Trij_offset = np.random.randint(0, 60*10)

d = utils.rollingWindow_from_df(df, wSize, columns=columns)
trajno = np.linspace(Trij_offset, d.shape[0]-1, nTraj).astype(int)
x = d[trajno, :, 0].T.cumsum(axis=0)
y = d[trajno, :, 1].T.cumsum(axis=0)

nTraj_perPlot = 10
cmap = plt.get_cmap('viridis', nTraj_perPlot)
i = 0
while (i < nTraj):
    if i % nTraj_perPlot == 0:
        fig, ax = plt.subplots(1, 1, figsize=(4, 4))
    ax.plot(x[:, i], y[:, i], alpha=0.5, c=cmap(i % nTraj_perPlot))
    ax.plot(0, 0, '.r')
    i += 1
```
